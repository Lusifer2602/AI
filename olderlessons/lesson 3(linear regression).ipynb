#REGRESSION

The term regression is used when you want to find relationship between variables
In ML & Statistical modelling, this relationship is used to predict the outcome of future events.

#Linear regression

It uses relation between data-points to draw a straight line through all of them.
This line can be used to predict future values.

```python
# Python has methods for finding a relationship between data-points and to draw a line of linear regression.
# linear regression starts with drawing a plot
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
# scipy's stats is used to draw linear regression through a plot
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] #an array is created

slope, intercept, r, p, std_err = stats.linregress(x, y) #this is a method that returns that key values of regression

def myfunc(x): #this function is created to use slope and intercept values to create a new value.
  return slope * x + intercept

mymodel = list(map(myfunc, x)) #this will-->Run each value of the x array through the function. This will result in a new array with new values for the y-axis:

plt.scatter(x, y) 
#upper and lower graphs will be integrated together to give us the exact linear regression of this graph
plt.plot(x, mymodel)
plt.show()
```

```python
from scipy import stats
import numpy as np
x=np.random.uniform(23, 100, 20)
y=np.random.uniform(10, 200, 20)

slope, intercept, r, p, std_err=stats.linregress(x, y)

def myfunction(x):
    return slope*x+intercept

model=list(map(myfunction, x))
plt.scatter(x,y)
plt.plot(x, model)
plt.show()
```

```python
# r in the linregress stands for relationship between the values of x adn y axis, because without relationship linear regression cannot be used to predict anything.
# The relationship (aka- coefficient of correlation) is denoted by 'r'

# for example
from scipy import stats
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```

```python
# now these values can be used to predict future values from the following 
from scipy import stats
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunction(x):
    return slope*x + intercept

#below is how we can use linear regression to predict values
print('value of y at x=10 may be : ', myfunction(10)) #here we have predicted value on the graph if the value of point on x is '10'
```
