Polynomial Regression

If data points do not fit under a linear regression curve, it might be ideal for polynomial regression.
Polynomial regression, like linear regression, uses the relationship between the variables x and y to find the best way to draw a line through the data points.

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

slope, intercept, r, p, std_err=stats.linregress(x,y)

def myfunc(x):
    return slope*x+intercept

mymodel=list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x,mymodel)
plt.show()
print("value for 'r' in this relationship is: ", r)
#here as we can see the value of r is so low hence a new type of regression i.e. will be tried to be applied on this dataset
```

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel=np.poly1d(np.polyfit(x,y,3)) #numpy has a method that lets us create polynomial models, this is what we did in this line

myline=np.linspace(1,22, 100) # this spevifies how the line will display, it starts at 1, ends at 22 and 100 is for how precisely we want the line to curve

plt.scatter(x,y) # draws points of dataset
plt.plot(myline, mymodel(myline)) #draws the polynomial regression line
plt.show()
```

R-squared (relationship for polynomial regression between x and y axis)
 r-squared is the term used for identifying how good a relationshi between x&y axis would be if polynomial regression is applied over it.

 r-squared value ranges from 0 to 1, where 0 means no relationship and 1 means 100% related. Python and sklearn module will compute this value for us, all one's gotta do is now feed it the values

```python
#for eg to check the r-squared in above dataset this is how we'll do it
import numpy as np
from sklearn.metrics import r2_score
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel=np.poly1d(np.polyfit(x,y,3)) #This line fits a polynomial regression model of degree 3 to the input and output datasets using the polyfit function from NumPy. The polyfit function returns the coefficients of the polynomial, which are then used to create a polynomial function mymodel using the poly1d function from NumPy
print("r-squared value :", r2_score(y, mymodel(x))) #This line calculates the r-squared value of the polynomial regression model using the r2_score function and prints it to the console. The r2_score function takes two arguments: the true output values y and the predicted output values mymodel(x).

#the output 0.94 shows the relationship is very good
```

HOW TO CHOOSE APPROPRIATE DEGREE OF POLYNOMIAL FOR A DATASET WHEN PERFORMING POLYNOMIAL REGRESSION ON IT?


**with the increasing degree of the polynomial, the complexity of the model also increases. Therefore the value of 'n' must be chosen precisely. If this value is low, then the model won't be able to fit the data properly, the model will overfit the data easily.**
